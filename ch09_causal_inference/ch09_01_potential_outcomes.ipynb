## 이론적 배경

$\tau_i = Y_i(1) - Y_i(0)$. ATE: $\tau = E[Y(1) - Y(0)]$. ATT: $E[Y(1) - Y(0)|W=1]$. 기본 문제: 각 개체의 $Y_i(1), Y_i(0)$을 동시에 관측할 수 없음

## 구현 가이드

```python
import numpy as np
np.random.seed(42)
n = 1000
X = np.random.normal(0, 1, n)
# 잠재 결과
Y0 = 2 + X + np.random.normal(0, 1, n)
Y1 = 4 + 1.5*X + np.random.normal(0, 1, n)
true_ate = (Y1 - Y0).mean()
# 무작위 배정
W = np.random.binomial(1, 0.5, n)
Y_obs = W*Y1 + (1-W)*Y0
naive_ate = Y_obs[W==1].mean() - Y_obs[W==0].mean()
print(f"참 ATE: {true_ate:.3f}")
print(f"무작위 배정 추정: {naive_ate:.3f}")
```
