## 이론적 배경

Beta-Binomial: $\theta|y \sim \text{Beta}(\alpha+y, \beta+n-y)$. Normal-Normal: $\mu|y \sim N(\frac{n\bar{y}/\sigma^2 + \mu_0/\tau^2}{n/\sigma^2 + 1/\tau^2}, \frac{1}{n/\sigma^2 + 1/\tau^2})$

## 구현 가이드

```python
import numpy as np
from scipy import stats
# Normal-Normal 켤레
mu0, tau0 = 0, 10  # 사전
sigma = 2  # 알려진 분산
data = np.random.normal(5, sigma, 30)
n, ybar = len(data), data.mean()
# 사후
tau_post = 1 / (n/sigma**2 + 1/tau0**2)
mu_post = tau_post * (n*ybar/sigma**2 + mu0/tau0**2)
print(f"사전: N({mu0}, {tau0**2})")
print(f"사후: N({mu_post:.3f}, {tau_post:.4f})")
print(f"MLE: {ybar:.3f}")
```
