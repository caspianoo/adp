## 문제 1 풀이

깁스 샘플링과 해밀토니안 MC 핵심 구현

```python
import numpy as np, matplotlib.pyplot as plt
np.random.seed(42)
print('깁스 샘플링과 해밀토니안 MC - 풀이 1')
# 핵심 구현 코드
("=== 깁스 샘플링과 해밀토니안 MC ===")

# 깁스 샘플링: 이변량 정규
rho = 0.8
n_samples = 10000
samples = np.zeros((n_samples, 2))
x, y = 0.0, 0.0

for i in range(n_samples):
    x = np.random.normal(rho * y, np.sqrt(1 - rho**2))
    y = np.random.normal(rho * x, np.sqrt(1 - rho**2))
    samples[i] = [x, y]

```

**결과 해석**: 깁스 샘플링과 해밀토니안 MC의 핵심 개념을 구현을 통해 확인했습니다.
