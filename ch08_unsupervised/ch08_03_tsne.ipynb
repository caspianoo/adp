## 이론적 배경

$p_{j|i} = \frac{\exp(-\|x_i - x_j\|^2/2\sigma_i^2)}{\sum_{k\neq i}\exp(-\|x_i - x_k\|^2/2\sigma_i^2)}$. Student-t: $q_{ij} = \frac{(1+\|y_i-y_j\|^2)^{-1}}{\sum_{k\neq l}(1+\|y_k-y_l\|^2)^{-1}}$. KL: $\min \sum p_{ij}\log\frac{p_{ij}}{q_{ij}}$

## 구현 가이드

```python
import numpy as np, matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
digits = load_digits()
X, y = digits.data, digits.target
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, perp in zip(axes, [5, 30, 100]):
    tsne = TSNE(n_components=2, perplexity=perp, random_state=42)
    X_2d = tsne.fit_transform(X)
    scatter = ax.scatter(X_2d[:,0], X_2d[:,1], c=y, cmap='tab10', s=5, alpha=0.7)
    ax.set_title(f'perplexity={perp}')
plt.colorbar(scatter); plt.tight_layout(); plt.show()
```
